```python
## Feature Engineering
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df.drop('Data',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Hours']=df['Arrival_Time'].str.split(':').str[0]
df['Minutes']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
df['Arrival_Hours']=df['Hours'].astype(int)
df['Arrival_Minutes']=df['Minutes'].astype(int)
```

```python
df.drop('Arrival_Hours',axis=1,inplace=True)
df.drop('Arrival_Minutes',axis=1,inplace=True)

```

```python
df.info()
```

```python
df['Dep_Hours']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_Hours']=df['Dep_Hours'].astype(int)
df['Dep_Minutes']=df['Dep_Minutes'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4,np.nan:1})
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_Hours']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Mins']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df['Duration_Mins'] = df['Duration_Mins'].fillna(0).astype(int)
```

```python
df[df['Duration_Mins'].isnull()]
```
